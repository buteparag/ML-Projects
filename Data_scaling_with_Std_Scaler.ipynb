```python
import numpy as np 
import pandas as pd 
import warnings 
warnings.filterwarnings ("ignore")
```

```python
df=pd.read_csv("cars.csv")
df.head()
```

```python
df["normalized-losses"].replace("?",np.nan,inplace=True)
```

```python
df["horsepower"].replace("?",np.nan,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
df[["horsepower","normalized-losses"]]=si.fit_transform(df[["horsepower","normalized-losses"]])
df
```

```python
catcol=df.select_dtypes(object).columns
catcol
```

```python
from sklearn.preprocessing import OrdinalEncoder 
oe=OrdinalEncoder() 
df[catcol]=oe.fit_transform(df[catcol])
df
```

**Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler 
ss=StandardScaler()
df.iloc[:,:-1]=ss.fit_transform(df.iloc[:,:-1])
df
```

**Standard Scaler brings all values in between Positive to Negative**
