```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings ("ignore")
```

```python
df=pd.read_csv("cars.csv")
df.head()
```

```python
df.info()
```

There are some missing values present in data.


```python
df["normalized-losses"].value_counts()
```

```python
df["horsepower"].value_counts()
```

There are missing values as "?" are present in above two columns.

So we need to replace this with nan since pandas can handle only nan as missing value.

```python
df["normalized-losses"].replace("?",np.nan,inplace=True)
df["horsepower"].replace('?',np.nan,inplace=True)
```

```python
df.head()
```

**All Missing values are now as NaN.**

```python
df.info()
```

Still you can see the data type as object instead of int64 or float64.

So we can change the datatype using astype method.

But here we will be using **simple imputer** which will not only change the data type but also will fill the missing values with mean.

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
df[["normalized-losses","horsepower"]]=si.fit_transform(df[["normalized-losses","horsepower"]])
```

```python
df.head()
```

```python
df.info()
```

**This way we can handle the missing value.**
