```python
import pandas as pd
import warnings 
warnings.filterwarnings ("ignore")
```

To read tsv file in pandas we use pd.read_table("Name of file with extension")

You can use this method if you don't know the extension of any file. later you can add seperator as per the output displayed.

```python
df1=pd.read_table("chiporders.tsv")
df1.head()
```

We can also use pd.read_table("Name of file with extension") for reading excel file by adding seperator as comma

```python
df2=pd.read_table("cars.csv")
df2.head()
```

Here in optput it is not displayed properly so we will now add seperator as comma.

```python
df2=pd.read_table("cars.csv",sep=",")
df2.head()
```

To read csv file you can either use pd.read_csv("Name of file")

**OR**

pd.read_table("Name of file with extension",sep=",")


```python
df3=pd.read_csv("titanic.csv")
df3.head()
```

```python
df3=pd.read_table("titanic.csv",sep=",")
df3.head()
```

To read json file (it is basically list of dictionaries or only dictionary)

we use pd.read_json("Name of file with extension")

```python
df4=pd.read_json("simple_records.json")
df4.head()
```

If error comes then use pd.read_json("Name of files with extension",lines=True)

```python
df4=pd.read_json("simple_records.json",lines=True)
df4.head()
```

```python
df5=pd.read_table("simple_records.json")
df5.head()
```

```python
df5=pd.read_table("simple_records.json",sep=":")
df5.head()
```

```python
colname=["name","age","grade"]
df6=pd.read_table("simple_records.json",sep=",",header=None,names=colname)
df6.head()
```

**Reading file from Web**

Here we don't know the extension so use pd.read_table method

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df7=pd.read_table(url)
df7.head()
```

Now we can see that here the seperator is comma 

```python
df7=pd.read_table(url,sep=",")
df7.head()
```

Again here we need to mention the column names as pandas taking first row as the  column header. 

```python
colname=["sepal length","sepal width","petal length","petal width","class"]
df7=pd.read_table(url,sep=",",header=None,names=colname)
df7.head()
```
