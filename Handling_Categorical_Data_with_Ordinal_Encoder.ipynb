```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings ("ignore")
```

```python
df=pd.read_csv("cars.csv")
df.head()
```

```python
df.info()
```

```python
df["normalized-losses"].replace("?",np.nan,inplace=True)
```

```python
df["horsepower"].replace("?",np.nan,inplace=True)
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
df[["normalized-losses","horsepower"]]=si.fit_transform(df[["normalized-losses","horsepower"]])
df.head()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

Now For creating the ML Model we need encode all the catergorical data.

This is done by:

1.Label Encoder (implemented to y i.e target/output only)

2.Ordinal Encoder (implemented to x i.e.features/Input only)

Here y is already numeric in nature.

but x has columns which are catergorical in nature. So we will use them for encoding.

```python
df.select_dtypes(object).columns
```

```python
catcol=df.select_dtypes(object).columns
catcol
```

**Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df[catcol]=oe.fit_transform(df[catcol])
df
```

Here the target variable is **contineous** i.e it becomes the **Regression problem.**

So we are using **Ordinal Encoding.**

In classification tasks where **target variable is categorical** we will bw using **label Encoding.**
