```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
df[["normalized-losses","horsepower"]]=si.fit_transform(df[["normalized-losses","horsepower"]])
```

**Spliting the data into X (Input) & Y (Output)**

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

**Outlier is the observation that lies away at abnormal distance from other values in random sample of population.** 

Outliers are the measure of variability.

They ruin the accuracy of the model. hence it is necessary to remove them.


```python
sns.boxplot(data=x,x=y)
plt.grid()
```

Here we have taken only price into consideration, but looking at dataset we need to consider the make also.

```python
sns.boxplot(data=x,x=y,y="make")
plt.grid()
```
